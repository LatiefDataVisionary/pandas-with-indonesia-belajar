## **Pandas 54: Menyimpan Data Frame sebagai file CSV | Python Pandas | Belajar Data Science | Indonesia**


```python
import pandas as pd
import numpy as np
```

Persiapan DataFrame

```python
data = {'nama': ['bayu', 'indra', 'devi', 'agni'],
        'jenis_kelamin': ['L', 'L', 'P', 'L'],
        'usia': [23, 21, 22, 25]}
df = pd.DataFrame(data)
df
```

Menyimpan DataFrame sebagai file `CSV`

```python
df.to_csv('data_peserta.csv')
```

```python
df = pd.read_csv('data_peserta.csv')
df
```

```python
df = pd.read_csv('data_peserta.csv')
df
```

```python
df = pd.read_csv('data_peserta.csv', index_col=0)
df
```

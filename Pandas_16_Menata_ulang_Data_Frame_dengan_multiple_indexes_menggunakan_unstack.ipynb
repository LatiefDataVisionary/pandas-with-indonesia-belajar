## **Pandas 16: Menata ulang Data Frame dengan multiple indexes menggunakan `unstack()` _ Data Science**

```python
import pandas as pd
import numpy as np
```

Persiapan DataFrame

```python
df = pd.read_csv("https://raw.githubusercontent.com/boedybios/kaggle_explorations/master/belajar_matplotlib_dasar/data/titanicfull.csv")
df.head()
```

DataFrame dengan multiple indexes dari hasil grouping

```python
df.groupby(['sex', 'pclass'])['survived'].mean().to_frame()
```

Menata ulang DataFrame dengan multiple indexes

```python
df.groupby(['sex', 'pclass'])['survived'].mean().unstack()
```

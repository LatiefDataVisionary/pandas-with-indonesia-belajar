## **Pandas 07: Menghapus (drop) missing values(NaN) pada Data Frame _ Python Pandas _ Belajar Data Science**

```python
import numpy as np
import pandas as pd
from pandas import testing

print(np.__version__)
print(pd.__version__)
```

Persiapan DataFrame

```python
# Membuat data frame random
# df = pd.DataFrame(np.random.rand(5, 4), columns=tuple('ABCD'))
df = pd.DataFrame(np.random.rand(10, 4), columns=tuple('ABCD'))

# Menambahkan kolom index
df.reset_index(inplace=True)

# Menambahkan missing value secara acak
df = df.mask(np.random.random(df.shape) < 0.2) # 20% data akan menjadi missing value (NaN)
df.head(50)
```

```python
df = df.rename(columns={'index': 'Z'})
df.head()
```

```python
df_backup = df.copy(deep=True)
```

Menghapus (drop) setiap kolom yang mengandung missing value

```python
df = df.dropna(axis='columns')
df.head()
```

Menghapus (drop) setiap baris yang mengandung missing values

```python
df = df_backup.copy(deep=True)
df = df.dropna(axis='rows')
df.head()
```

Persentase missing values untuk tiap kolom

```python
df = df_backup.copy(deep=True)
df.isna().mean()
```

Mengapus (drop) setiap kolom yang mengandung missing values berdasarkan treshold

```python
treshold = len(df) * 0.9
df = df.dropna(thresh=treshold, axis='columns')
df.head()
```

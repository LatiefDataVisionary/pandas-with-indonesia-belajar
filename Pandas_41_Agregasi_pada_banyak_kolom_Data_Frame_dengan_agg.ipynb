## **Pandas 41: Agregasi pada banyak kolom Data Frame dengan agg() | Python Pandas | Data Science**


```python
import pandas as pd
import numpy as np
```

Persiapan DataFrame

```python
df = pd.read_csv('https://raw.githubusercontent.com/LatiefDataVisionary/pandas-with-indonesia-belajar/master/datasets/titanicfull.csv')
df.head()
```

Menerapkan agregasi pada sejumlah kolom dengan `agg()`

```python
df.groupby('pclass').agg({'pclass': 'count',
                          'age': ['mean', "max"],
                          'survived': 'mean'})
```

```python
result = df.groupby('pclass').agg({'pclass': 'count',
                                     'age': ['min', 'mean', "max"],
                                     'survived': 'mean'})
```

```python
result['age'] = result['age'].round(2)  # Membulatkan semua kolom di dalam level 'age'
result['survived'] = result['survived'].round(2)  # Membulatkan kolom 'survived'
```

```python
result[('age', 'mean')] = result[('age', 'mean')].round(2)
result[('age', 'max')] = result[('age', 'max')].round(2)
result
```

```python
df.groupby('pclass').agg(n_pass=('pclass', 'count'),
                         avg_age=('age', 'mean'),
                         max_age=('age', 'max'),
                         survival_rate=('survived', 'mean'))
```

```python
df.groupby('pclass').agg(
    n_pass=('pclass', 'count'),
    female_pass=('sex', lambda x: (x == 'female').sum()),  # Menghitung jumlah 'female'
    male_pass=('sex', lambda x: (x == 'male').sum()),  # Menghitung jumlah 'male'
    min_age=('age', 'min'),
    avg_age=('age', 'mean'),
    max_age=('age', 'max'),
    survival_rate=('survived', 'mean')
)
```

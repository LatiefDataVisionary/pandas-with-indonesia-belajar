## **Pandas 08: Memeriksa kesamaan antar kolom (Series) pada Data Frame _ Python Pandas _ Data Science**

```python
import pandas as pd
import numpy as np
```

Mengenal Pandas Series

```python
data = {'A': [15, 15, 18, np.nan, 12],
        'B': [15, 15, 18, np.nan, 12]}
df = pd.DataFrame(data)
df
```

Mengenal Pandas Series

```python
df['A']
```

```python
type(df['A'])
```

```python
type(df)
```

Memeriksa kesamaan dengan operator `==`

```python
df['A'] == df['B']
```

Memeriksa kesamaan dengan method `equals()`

```python
df['A'].equals(df['B'])
```

Memeriksa kesamaan antar dua DataFrame

```python
df1 = df.copy(deep=True)
df1.equals(df)
```

```python
df == df1
```

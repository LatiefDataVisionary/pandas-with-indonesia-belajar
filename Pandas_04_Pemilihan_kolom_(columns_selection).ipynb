## **Pandas 04: Pemilihan kolom (columns selection) pada Pandas Data Frame berdasarkan tipe data**

```python
import pandas as pd
import numpy as np
```

Persiapan Data Frame

```python
n_rows = 5
n_cols = 2
cols = ['bil_pecahan', 'bil_bulat']

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)),
                  columns=cols)
df['bil_pecahan'] = df['bil_pecahan'].astype('float')

## Replace pd.util.testing.makeDateIndex with pd.date_range
# df.index = pd.util.testing.makeDateIndex(n_rows, freg='H')
df.index = pd.date_range('2023-11-21', periods=n_rows, freq='H') # H artinya Hour
df = df.reset_index()

df['teks'] = list('ABCDE')
df
```

```python
df.dtypes
```

Mmeilih kolom bertipe data numerik

```python
df.select_dtypes(include='number')  # Memilih kolom bertipe data numerik
```

```python
df.select_dtypes(include='float')  # Memilih kolom bertipe data float
```

```python
df.select_dtypes(include='int')   # Memilih kolom bertipe data int
```

Memilih kolom bertipe data `string` atau `object`

```python
df.select_dtypes(include='object')
```

Mmeilih kolom bertipe data `datetime`

```python
df.select_dtypes(include='datetime')
```

Memilih kolom dengan kombinasi tipe data

```python
df.select_dtypes(include=['number', 'object'])  # Menggunakan list []
```

```python
df.select_dtypes(include=['object', 'int', 'float', 'datetime'])
```

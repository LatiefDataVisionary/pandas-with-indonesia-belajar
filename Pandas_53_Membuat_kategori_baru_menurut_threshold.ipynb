## **Pandas 53: Membuat kategori baru menurut threshold (ambang batas) | Python Pandas | Data Frame | Data Science**


```python
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

Persiapan DataFrame

```python
d = {'hobi':['jogging', 'mancing', 'renang',
             'mancing', 'mancing', 'baca',
             'baca', 'mancing', 'fotografi',
             'mancing', 'camping']}

df = pd.DataFrame(d)
df
```

Membuat kategori baru berdasarkan treshold (ambang batas)

```python
df['hobi'].value_counts().to_frame()
```

```python
persentase = df['hobi'].value_counts(normalize=True).to_frame()
persentase
```

```python
threshold = 0.1 # utk 10%
hobi_lain = persentase[persentase['proportion'] < threshold].index.to_list()
df['hobi'] = df['hobi'].replace(hobi_lain, 'lainnya')
```

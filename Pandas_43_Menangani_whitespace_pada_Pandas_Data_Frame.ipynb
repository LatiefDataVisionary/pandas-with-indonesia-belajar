## **Pandas 43: Menangani whitespace pada Pandas Data Frame | Python Pandas | Belajar Data Science**


```python
import pandas as pd
import numpy as np
```

Persiapan DataFrame

```python
data = {'nim':['10', '11', '12', '13', ' '],
        'nama': ['adi', ' ', 'tejo', ' ', 'bejo']}
df = pd.DataFrame(data)
df
```

```python
df.info()
```

```python
df = df.replace(r'^\s*$', np.nan, regex=True)
df
```

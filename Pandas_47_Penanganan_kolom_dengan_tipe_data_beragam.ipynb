## **Pandas 47: Penanganan kolom dengan tipe data beragam (mixed data types) | Python Pandas | Belajar Data Science**


```python
import pandas as pd
import numpy as np
```

Persiapan DataFrame

```python
d = {'nama': ['bejo', 'tejo', 'wati', 'tiwi', 'cecep'],
     'ipk': [2, '3', 3, 2.75, '3.75']}
df = pd.DataFrame(d)
df
```

Deteksi dan penanganan kolom dengan tipe data beragam (mixed data types)

```python
df.dtypes.to_frame()
```

```python
df['ipk'].apply(type)
```

```python
df['ipk'].apply(type).value_counts().to_frame()
```

```python
df['ipk'] = df['ipk'].astype('float')
```

```python
df['ipk'].apply(type).value_counts()
```

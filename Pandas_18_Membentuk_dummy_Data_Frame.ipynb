Membentuk DataFrame dengan memanfaatkan `pandas.util.testing`

```python
# pd.util.testing.makeDataFrame().head()

df1 = pd.DataFrame(np.random.randn(5, 4), columns=['A', 'B', 'C', 'D']) # setara dengan fungsi lama, membuat DataFrame acak
print("DataFrame setara makeDataFrame():\n", df1.head())
```

```python
# pd.util.testing.makeMixedDataFrame().head()

df2 = pd.DataFrame({'A': [0., 1., 2., 3., 4.],
                   'B': pd.Categorical(["test", "train", "test", "train", "test"]),
                   'C': ['foo1', 'foo2', 'foo3', 'foo4', 'foo5'],
                   'D': pd.Timestamp('20090101')}) # Setara dengan fungsi lama, membuat DataFrame dengan kolom tipe data campuran.
print("\nDataFrame setara makeMixedDataFrame():\n", df2.head())
```

```python
# pd.util.testing.makeTimeDataFrame().head()

df3 = pd.DataFrame(index=pd.date_range('20090101', periods=5)) # Setara dengan fungsi lama, membuat DataFrame dengan index berupa deret waktu.
df3['A', 'B', 'C', 'Latief Ganteng'] = np.random.randn(5)
print("\nDataFrame setara makeTimeDataFrame():\n", df3.head())
```

```python
df4 = pd.DataFrame(index=pd.date_range('20090101', periods=5))
# Buat list berisi nama kolom
nama_kolom = ['A', 'B', 'C', 'Latief Ganteng']

# Buat data acak untuk kolom
data_acak = np.random.randn(5, len(nama_kolom))  # 5 baris, 4 kolom

# Tetapkan data ke kolom DataFrame
df4[nama_kolom] = data_acak

# Tampilkan DataFrame
print("\nDataFrame dengan 4 kolom:\n", df3.head())
```

```python
# pd.util.testing.makeMissingDataFrame().head()

df4 = pd.DataFrame(np.random.randn(5, 4), columns=['A', 'B', 'C', 'D']) # setara dengan fungsi lama, membuat DataFrame acak
df4.iloc[2, 2] = np.nan # setara dengan fungsi lama, menambahkan nilai NaN ke dataframe
print("\nDataFrame setara makeMissingDataFrame():\n", df4.head())
```

## **Pandas 29: Random Sampling pada Pandas Data Frame _ Python Pandas _ Belajar Data Science**

```python
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

Persiapan DataFrame

```python
d = {'col_1':[1,2,3,4,5],
     'col_2':[10,20,30,40,50]}
df = pd.DataFrame(d)
df
```

Random smapling with/without replacement

```python
df.sample(n=4, replace=False, random_state=0)
```

```python
df.sample(n=4, replace=True, random_state=0)
```

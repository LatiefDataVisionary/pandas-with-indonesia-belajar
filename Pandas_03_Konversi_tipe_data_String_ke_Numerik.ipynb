## **Pandas 03: Konversi tipe data String ke Numerik pada kolom Data Frame _ Python Pandas**

```python
import pandas as pd
import numpy as np
```

Persiapan Data Frame

```python
data = {'col1': ['1', '2', '3', 'teks'],
        'col2': ['1', '2', '3', '4']}

df = pd.DataFrame(data)
df
```

```python
df.dtypes
```

Konversi tipe data dnegan fungsi `astype()`

```python
df_x = df.astype({'col2': 'int'})
df_x
```

```python
df_x.dtypes
```

```python
df_x = df.astype({'col2': 'float'})
df_x
```

```python
df_x.dtypes
```

Konversi tipe data dengan fungsi `to_numeric()`

```python
df.apply(pd.to_numeric, errors='coerce') # coerce menghasilkan nilia NaN/missing value jika string dibuat ke numerik
```

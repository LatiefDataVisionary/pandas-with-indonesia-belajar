## **Pandas 09: Membagi Data Frame menjadi dua secara acak _ Python Pandas _ Belajar Data Science**

```python
import numpy as np
import pandas as pd

print(np.__version__)
print(pd.__version__)
```

Persiapan DataFrame

```python
n_rows = 10
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1, 10, size=(n_rows, n_cols)), columns=cols)
df
```

Membagi DataFrame menjadi dua scera acak berdasarkan proporsi tertentu

```python
df.shape
```

```python
proporsi = 0.7
df_1 = df.sample(frac=proporsi)
df_2 = df.drop(df_1.index)

print(f'df_1 shape: {df_1.shape}')
print(f'df_2 shape: {df_2.shape}')
```

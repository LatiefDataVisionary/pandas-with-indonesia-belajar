## **Pandas 14: Menggabungkan dua Pandas Data Frame _ Python Pandas _ Belajar Data Science _ Indonesia**

```python
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

Persiapan DataFrame

```python
n_rows = 5
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)),
                  columns=cols)
df
```

```python
df1 = df.copy(deep=True)
df1 = df1.drop([1, 4])  # Membuang index 1 dan 4
df1
```

```python
df2 = df.copy(deep=True)
df2 = df2.drop([0, 3])  # Membuang indeks 0 dan 3
df2
```

Menggabungkan dua DataFrame

```python
df_inner = pd.merge(df1, df2, how='inner')
df_inner
```

```python
df_outer = pd.merge(df1, df2, how='outer')
df_outer
```

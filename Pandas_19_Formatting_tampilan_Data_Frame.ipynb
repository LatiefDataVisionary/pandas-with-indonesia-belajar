## **Pandas 19: Formatting tampilan Data Frame _ Python Pandas _ Data Science _ Bahasa Indonesia**

```python
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

Persiapan DataFrame

```python
n_rows = 5
n_cols = 2
cols = ['omset', 'operasional']

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)),
                  columns=cols)
df
```

```python
df['omset'] = df['omset'] * 100_000
df['operasional'] = df['operasional'] * 10_000
df
```

```python
# df.index = pd.util.testing.makeDateIndex(n_rows, freg='D')
df.index = pd.date_range(start='2024-01-01', periods=n_rows, freq='D')
df = df.reset_index()
df = df.rename(columns={'index': 'tanggal'})
df
```

Melakukan formatting tampilan DataFrame

```python
formatku = {'tanggal': '{:%d/%m/%y}',
            'operasional': 'Rp {:.2f}',
            'omset': 'Rp {:.2f}'}

laporan = df.style.format(formatku)
laporan
```

```python
type(laporan)
```

```python
# laporan.hide_index()
laporan.hide(axis="index") # Or laporan.hide() for hiding the index
```

```python
laporan.set_caption('Data Omset dan Operasional')
```

```python
laporan.highlight_min('omset', color='lightyellow')
laporan.highlight_max('omset', color='lightgreen')

laporan.highlight_min('operasional', color='lightblue')
laporan.highlight_max('operasional', color='grey')
```
